```python

#finds apartments inside radius r of given point
def neighbouring_points_within_r(point, r, tree_structure):
    indices = tree_structure.query_ball_point(point, r)
    return filtrirana_kupljena_stanovanja.iloc[indices]


# finds k-nearest apartments of given point
def k_neighbouring_points(point, k, tree_structure):
    dist, indices = tree_structure.query(point, k)
    return filtrirana_kupljena_stanovanja.iloc[indices]
```

```python
from dateutil.relativedelta import relativedelta
tree = spatial.cKDTree(filtrirana_kupljena_stanovanja[['position_x','position_y']].to_numpy())

def comparison_function_apartments(apartment,trees = tree):
    #apartments in radius 250m
    apartments_candidates = neighbouring_points_within_r([float(apartment.position_x.values[0]),float(apartment.position_y.values[0])],250,trees)
    #returns k most comparable apartments
    std = apartments_candidates['CENA_M_2'].std()
    mean = apartments_candidates['CENA_M_2'].mean()
    #filters edges
    removed_noise_data =  apartments_candidates[(apartments_candidates['CENA_M_2'] < (2* std + mean)) & (apartments_candidates['CENA_M_2'] > (mean-2*std))]
    #creates column for similarity metrics and distance
    #calculates distances from apartment
    distance = list(map(lambda l:((float(l[0])-float(apartment.position_x.values[0]))**2+(float(l[1])-float(apartment.position_y.values[0]))**2)**0.5,removed_noise_data[['position_x','position_y']].to_numpy()))
    removed_noise_data = removed_noise_data.assign(DISTANCE = distance)
    removed_noise_data = removed_noise_data.assign(SIMILARITY = [0 for i in range(len(removed_noise_data))])
    #Define weights
    weight_same_street = 5.0
    weight_same_building = 5.0
    weight_proximity = 1.0
    weight_valuation_zone = 4.0
    weight_same_usage_type = 1.0
    weight_size = 3.0
    weight_transaction_date = 4.0
    weight_year_build = 3.0
    #similarity calculator
    #1. same street criteria
    removed_noise_data.loc[removed_noise_data.index[(removed_noise_data['Ulica'] == apartment['Ulica'].values[0]) & (removed_noise_data['ST_NADSTROPJA'].astype(float) <= float(apartment['ST_NADSTROPJA'].values[0])+3) & (removed_noise_data['ST_NADSTROPJA'].astype(float) >= float(apartment['ST_NADSTROPJA'].values[0])-3) & (removed_noise_data['LETO_IZG_STA'].astype(float) <= float(apartment['LETO_IZG_STA'].values[0])+3) & (removed_noise_data['LETO_IZG_STA'].astype(float) >= float(apartment['LETO_IZG_STA'].values[0])-3)],'SIMILARITY'] -= 10*weight_same_street
    removed_noise_data.loc[:,'SIMILARITY'] += 10*weight_same_street
    #2. same building criteria
    removed_noise_data.loc[removed_noise_data.index[removed_noise_data['STA_SID'] != apartment['STA_SID'].values[0]],'SIMILARITY'] += 1.0 *weight_same_building
    #3. Proximity criteria
    removed_noise_data.loc[removed_noise_data.index[(removed_noise_data['DISTANCE'] > 50.0) & (removed_noise_data['DISTANCE'] <= 200.0)],'SIMILARITY'] += 2.0 *weight_proximity
    removed_noise_data.loc[removed_noise_data.index[removed_noise_data['DISTANCE'] > 200],'SIMILARITY'] += 5.0 *weight_proximity
    #4. Valuation zone criteria
    removed_noise_data.loc[removed_noise_data.index[removed_noise_data['ST_RAVNI'] == apartment.ST_RAVNI.values[0] ],'SIMILARITY'] += 0 *weight_valuation_zone - 50.0*weight_valuation_zone
    removed_noise_data.loc[removed_noise_data.index[(removed_noise_data['ST_RAVNI'] == apartment.ST_RAVNI.values[0]-1) | (removed_noise_data['ST_RAVNI'] == apartment.ST_RAVNI.values[0]+1) ],'SIMILARITY'] += 2.0*weight_valuation_zone-weight_valuation_zone*50.0
    removed_noise_data.loc[removed_noise_data.index[(removed_noise_data['ST_RAVNI'] == apartment.ST_RAVNI.values[0]-2) | (removed_noise_data['ST_RAVNI'] == apartment.ST_RAVNI.values[0]+2) ],'SIMILARITY'] += 10.0*weight_valuation_zone-50*weight_valuation_zone
    removed_noise_data.loc[:,'SIMILARITY'] += 50.0*weight_valuation_zone
    #5. Usage Type criteria
    #removed_noise_data.loc[removed_noise_data.index[removed_noise_data['Dejanska raba dela stavbe'] != apartment['Dejanska raba dela stavbe'].values[0]],'SIMILARITY'] += 5.0 *weight_same_usage_type
    # 6. Size criteria
    size_difference = abs((removed_noise_data['NETO_TLORIS_POV_DST'].astype(float)-float(apartment.NETO_TLORIS_POV_DST.values[0])))
    removed_noise_data.loc[(size_difference > 0.10 * float(apartment.NETO_TLORIS_POV_DST.values[0])) & (size_difference <= 0.15 * float(apartment.NETO_TLORIS_POV_DST.values[0])),'SIMILARITY'] += 5.0 * weight_size
    removed_noise_data.loc[(size_difference > 0.15 * float(apartment.NETO_TLORIS_POV_DST.values[0])) & (size_difference <= 0.25 * float(apartment.NETO_TLORIS_POV_DST.values[0])),'SIMILARITY'] += 10.0 * weight_size
    removed_noise_data.loc[(size_difference > 0.25 * float(apartment.NETO_TLORIS_POV_DST.values[0])) & (size_difference <= 0.50 * float(apartment.NETO_TLORIS_POV_DST.values[0])),'SIMILARITY'] += 15.0 * weight_size
    removed_noise_data.loc[size_difference > 0.50 * float(apartment.NETO_TLORIS_POV_DST.values[0]),'SIMILARITY'] += 100.0 * weight_size
    # 7. Transaction date
    removed_noise_data.loc[removed_noise_data.index[(abs((removed_noise_data['Datum sklenitve pogodbe']- apartment['Datum sklenitve pogodbe'].values[0])/np.timedelta64(1, "D")) < 365 )& (abs((removed_noise_data['Datum sklenitve pogodbe']- apartment['Datum sklenitve pogodbe'].values[0])/np.timedelta64(1, "D")) >= 183)],'SIMILARITY'] += 10.0 *weight_transaction_date
    removed_noise_data.loc[removed_noise_data.index[(abs((removed_noise_data['Datum sklenitve pogodbe']- apartment['Datum sklenitve pogodbe'].values[0])/np.timedelta64(1, "D")) < 365*2 )& (abs((removed_noise_data['Datum sklenitve pogodbe']- apartment['Datum sklenitve pogodbe'].values[0])/np.timedelta64(1, "D")) >= 365)],'SIMILARITY'] += 30.0 *weight_transaction_date
    removed_noise_data.loc[removed_noise_data.index[abs((removed_noise_data['Datum sklenitve pogodbe']- apartment['Datum sklenitve pogodbe'].values[0])/np.timedelta64(1, "D")) >= 365*2],'SIMILARITY'] += 40.0 *weight_transaction_date
    'Leto izgradnje dela stavbe'
    #8. Year built
    # intervals for apartments
    #<1945
    #[1946,1965]
    #[1966,1981]
    #[1982,1977]
    #[1998,2008]
    #>2008
    removed_noise_data.loc[removed_noise_data.index[abs(removed_noise_data['LETO_IZG_STA'].astype(float)- float(apartment['LETO_IZG_STA'].values[0])) > 20],'SIMILARITY'] += (abs(removed_noise_data.loc[removed_noise_data.index[abs(removed_noise_data['LETO_IZG_STA'].astype(float)- float(apartment['LETO_IZG_STA'].values[0])) > 20]]['LETO_IZG_STA'].astype(float)-float(apartment['LETO_IZG_STA'].values[0])) + 50)*weight_year_build
    removed_noise_data.loc[removed_noise_data.index[abs(removed_noise_data['LETO_IZG_STA'].astype(float)- float(apartment['LETO_IZG_STA'].values[0])) <= 20],'SIMILARITY'] += abs(removed_noise_data.loc[removed_noise_data.index[abs(removed_noise_data['LETO_IZG_STA'].astype(float)- float(apartment['LETO_IZG_STA'].values[0])) <= 20]]['LETO_IZG_STA'].astype(float)-float(apartment['LETO_IZG_STA'].values[0]))*weight_year_build
    #normalization
    removed_noise_data['SIMILARITY'] = removed_noise_data['SIMILARITY'].map(lambda x: int((1 - 2/np.pi *np.arctan(x*0.00085))*100))
    #returns sorted by  similarity
    return removed_noise_data.sort_values(by = ['SIMILARITY'],ascending= False,ignore_index = True)
```

```python
def evaluate_apartment(apartment):
    primerljivi = comparison_function_apartments(apartment)
    primerljivi = primerljivi[primerljivi['SIMILARITY'] > 90]

    return sum(primerljivi['CENA_M_2']*primerljivi['SIMILARITY']/100)/sum(primerljivi['SIMILARITY']/100)
```
